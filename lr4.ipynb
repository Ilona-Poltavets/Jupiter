```python
import numpy as np
import pandas as pd

# 1. Перетворення словника на серію Pandas
data_dict = {'a': 100, 'b': 200, 'c': 300, 'd': 400, 'e': 800}
series_from_dict = pd.Series(data_dict)
print("1. Перетворення словника на серію Pandas:")
print(series_from_dict)
print()

# 2. Перетворення масиву NumPy у серію Pandas
data_array = np.array([10, 20, 30, 40, 50])
series_from_array = pd.Series(data_array)
print("2. Перетворення масиву NumPy у серію Pandas:")
print(series_from_array)
print()

# 3. Додавання деяких даних до існуючої серії Pandas
existing_series = pd.Series([100, 200, 'python', '300,12', 400], index=[0, 1, 2, 3, 4])
additional_data = pd.Series([500, 'php'], index=[0, 1])
new_series = pd.concat([existing_series, additional_data])
print("3. Додавання деяких даних до існуючої серії Pandas:")
print(new_series)

print()

# 4. Створення DataFrame з вказаних даних та виконання завдань
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

# Створення DataFrame
df = pd.DataFrame(data, index=labels)

# a) Вибір лише стовпців «тварина» та «вік»
print("a) Вибір лише стовпців «тварина» та «вік»:")
print(df[['animal', 'age']])
print()

# b) Вибір даних в рядках [3, 4, 8] і в стовпцях ['animal', 'age']
print("b) Вибір даних в рядках [3, 4, 8] і в стовпцях ['animal', 'age']:")
print(df.loc[df.index[[3, 4, 8]], ['animal', 'age']])
print()

# c) Вибір лише тих рядків, де кількість відвідувань перевищує 3
print("c) Вибір лише тих рядків, де кількість відвідувань перевищує 3:")
print(df[df['visits'] > 3])
print()

# d) Вибір рядків, де тварина — кішка, а вік — менше 3-х
print("d) Вибір рядків, де тварина — кішка, а вік — менше 3-х:")
print(df[(df['animal'] == 'cat') & (df['age'] < 3)])
print()

# e) Вибір рядків віком від 2 до 4 (включно)
print("e) Вибір рядків віком від 2 до 4 (включно):")
print(df[(df['age'] >= 2) & (df['age'] <= 4)])
print()

# f) Обчислення суми всіх відвідувань у df
print("f) Обчислення суми всіх відвідувань у df:")
print(df['visits'].sum())
print()

# g) Додавання нового рядка 'k' до df з вибором значень для кожного стовпця,
#    а потім видалення цього рядка
df.loc['k'] = ['lion', 1.5, 1, 'no']
print("g) Додавання нового рядка 'k' до df з вибором значень для кожного стовпця:")
print(df)
print("Видалення рядка 'k' з df:")
df = df.drop('k')
print(df)
```
