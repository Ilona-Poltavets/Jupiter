```python
import numpy as np
import matplotlib.pyplot as plt

fsize = 12

plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = 'Times New Roman'
plt.rcParams['axes.titlesize'] = fsize
plt.rcParams['axes.labelsize'] = fsize
plt.rcParams['xtick.labelsize'] = fsize
plt.rcParams['ytick.labelsize'] = fsize
plt.rcParams['legend.fontsize'] = fsize

# Масив координат
x = np.linspace(0.0, 2*np.pi, 50)

# Створення вікна малюнка та панелей
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Графік синуса
axs[0, 0].plot(x, np.sin(x), 'k-', label='sin(x)')
axs[0, 0].set_title('(а)')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('sin(x)')
axs[0, 0].set_xlim(0, 2*np.pi)
axs[0, 0].set_ylim(-1, 1)

# Графік косинуса
axs[0, 1].plot(x, np.cos(x), 'k--', label='cos(x)')
axs[0, 1].set_title('(б)')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('cos(x)')
axs[0, 1].set_xlim(0, 2*np.pi)
axs[0, 1].set_ylim(-1, 1)

# Графік квадрата косинуса
axs[1, 0].plot(x, np.cos(x)**2, 'k-.', label='cos^2(x)')
axs[1, 0].set_title('(в)')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('cos^2(x)')
axs[1, 0].set_xlim(0, 2*np.pi)
axs[1, 0].set_ylim(0, 1)

# Графік функції x^0.15
axs[1, 1].plot(x, x**0.15, 'grey', label='$x^{0.15}$')
axs[1, 1].set_title('(г)')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('$x^{0.15}$')
axs[1, 1].set_xlim(0, 2*np.pi)

# Зберігаємо малюнок
plt.tight_layout()
plt.savefig("fig1_modified.png", dpi=300)
plt.show()
```
