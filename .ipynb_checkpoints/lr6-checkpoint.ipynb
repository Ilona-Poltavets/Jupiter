```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#  
titanic_data = pd.read_csv("titanic_train.csv")

# 1.   Cabin      
titanic_data.drop(columns=['Cabin'], inplace=True)
titanic_data.dropna(inplace=True)

#   
sns.pairplot(titanic_data[['Age', 'Fare', 'Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked', 'Survived']])
plt.show()

# 2.  boxplot          (Pclass)
sns.boxplot(x='Pclass', y='Fare', data=titanic_data)
plt.show()

# C  Fare_no_out
fare_mean_std_by_pclass = titanic_data.groupby('Pclass')['Fare'].agg(['mean', 'std'])
titanic_data['Fare_no_out'] = titanic_data.apply(lambda x: x['Fare'] if (x['Fare'] <= fare_mean_std_by_pclass.loc[x['Pclass'], 'mean'] + 2 * fare_mean_std_by_pclass.loc[x['Pclass'], 'std']) and (x['Fare'] >= fare_mean_std_by_pclass.loc[x['Pclass'], 'mean'] - 2 * fare_mean_std_by_pclass.loc[x['Pclass'], 'std']) else None, axis=1)

# 3.  countplot   
sns.countplot(x='Survived', hue='Sex', data=titanic_data)
plt.show()

# 4.  countplot    
sns.countplot(x='Survived', hue='Pclass', data=titanic_data)
plt.show()

# 5.       
titanic_data['Age_group'] = pd.cut(titanic_data['Age'], bins=[0, 30, 60, titanic_data['Age'].max()], labels=['Young', 'Adult', 'Elderly'])
sns.countplot(x='Survived', hue='Age_group', data=titanic_data)
plt.show()
```
