```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline

data = pd.read_csv("weights_heights.csv", index_col="Index")

data.plot(y="Height", kind="hist", color="red", title="Height (inch.) distribution");

def make_bmi(height_inch, weight_pound):
    METER_TO_INCH, KILO_TO_POUND = 39.37, 2.20462
    return (weight_pound / KILO_TO_POUND) / (height_inch / METER_TO_INCH) ** 2
data["BMI"] = data.apply(lambda row: make_bmi(row["Height"], row["Weight"]), axis=1)

# Task 2
# Перегляд перших 5 записів
print(data.head())

# Гістограма розподілу ваги
data.plot(y="Weight", kind="hist", color="green", title="Weight Distribution")
plt.xlabel("Weight (lbs)")
plt.ylabel("Frequency")
plt.show()

# Task 3
# Побудова попарних залежностей
sns.pairplot(data)
plt.show()

# Task 4
# Створення нової ознаки 'weight_category'
def weight_category(weight):
    if weight < 120:
        return 1
    elif weight >= 150:
        return 3
    else:
        return 2

data['weight_category'] = data['Weight'].apply(weight_category)

# Ящик з вусами для залежності зростання від вагової категорії
sns.boxplot(x='weight_category', y='Height', data=data)
plt.xlabel('Weight Category')
plt.ylabel('Height')
plt.show()

# Task 5
# Діаграма розсіювання залежно від ваги
data.plot(x='Weight', y='Height', kind='scatter', title='Height vs. Weight')
plt.xlabel('Weight (lbs)')
plt.ylabel('Height (inches)')
plt.show()
```
